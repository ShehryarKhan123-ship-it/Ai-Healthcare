### Preprocssing the Dataset

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
le = LabelEncoder()
df['label'] = le.fit_transform(df['diagnosis'])

# Train-test split
train_texts, val_texts, train_labels, val_labels = train_test_split(df['query'].tolist(), df['label'].tolist(), test_size=0.1, random_state=42)
```

```python
from transformers import AutoTokenizer

# Use BioBERT from Hugging Face
model_name = "dmis-lab/biobert-base-cased-v1.1"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Tokenize
train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=128)
val_encodings = tokenizer(val_texts, truncation=True, padding=True, max_length=128)
```

```python
import torch

class MedQueryDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {k: torch.tensor(v[idx]) for k, v in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = MedQueryDataset(train_encodings, train_labels)
val_dataset = MedQueryDataset(val_encodings, val_labels)
```

### 📦 Creating a Custom Dataset for PyTorch Model Training
This cell defines a custom PyTorch Dataset class called MedicalDataset, which prepares tokenized biomedical data and labels for model training and evaluation.

Key components:

__init__(): Accepts encodings (tokenized inputs) and labels, storing them for use during iteration.

__getitem__(): Retrieves a single data point by index idx. Each item includes:

Tokenized inputs (e.g., input_ids, attention_mask) converted into PyTorch tensors.

The corresponding label, also converted into a tensor and added under the key 'labels'.

__len__(): Returns the total number of samples in the dataset.

Finally, two dataset instances are created:

train_dataset for training

test_dataset for evaluation

```python
from transformers import AutoModelForSequenceClassification

num_labels = len(le.classes_)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)
```

### Training the model

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./biobert-medical",
    save_strategy="epoch",
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    num_train_epochs=2,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=100
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset
)

trainer.train()
```

```python
trainer.save_model("./biobert-medical")
```

```python
 trainer.evaluate()

from sklearn.metrics import classification_report

# Predictions
preds_output = trainer.predict(val_dataset)
preds = torch.argmax(torch.tensor(preds_output.predictions), axis=1)

# Report
print(classification_report(val_labels, preds, target_names=le.classes_))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(val_labels, preds)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

```python
import torch
import gradio as gr
from gradio.themes.utils import colors, sizes, fonts
import socket
from contextlib import closing
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import joblib

# Load tokenizer and model
model_path = "./biobert-medical"
tokenizer = AutoTokenizer.from_pretrained("dmis-lab/biobert-base-cased-v1.1")
model = AutoModelForSequenceClassification.from_pretrained(model_path)
model.eval()

import pickle

with open("label_encoder.pkl", "rb") as f:
    label_encoder = pickle.load(f)

# Load label encoder
label_encoder = joblib.load("label_encoder.pkl")

# Custom professional medical theme
med_theme = gr.themes.Default(
    primary_hue=colors.blue,
    secondary_hue=colors.teal,
    neutral_hue=colors.gray,
    spacing_size=sizes.spacing_md,
    radius_size=sizes.radius_md,
    text_size=sizes.text_md,
    font=fonts.GoogleFont("Poppins")
).set(
    button_primary_background_fill="linear-gradient(90deg, #1a73e8 0%, #0d47a1 100%)",
    button_primary_text_color="#ffffff",
    button_primary_background_fill_hover="linear-gradient(90deg, #0d47a1 0%, #1a73e8 100%)",
    shadow_spread="6px",
    block_background_fill="#f9f9f9",
    block_label_background_fill="#0d47a1",
    block_label_text_color="#ffffff",
    block_title_text_color="#1a73e8"
)

def find_free_port(start_port=7860):
    with closing(socket.socket(socket.AF_INET, socket.SOCK_STREAM)) as s:
        current_port = start_port
        while current_port < 65535:
            try:
                s.bind(('', current_port))
                s.close()
                return current_port
            except OSError:
                current_port += 1
        raise OSError("No free ports available")

def predict(symptoms):
    try:
        if not symptoms.strip():
            return (
                "## DIAGNOSTIC REPORT\n\n\u26a0\ufe0f Please describe your symptoms",
                {"No input": 0.0},
                "### CLINICAL RECOMMENDATION\n\n*No input provided*"
            )

        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        model.to(device)

        inputs = tokenizer(symptoms, return_tensors="pt", truncation=True, padding=True, max_length=128).to(device)
        with torch.no_grad():
            outputs = model(**inputs)
            logits = outputs.logits
            predicted_class_id = torch.argmax(logits, dim=1).item()
            predicted_label = label_encoder.inverse_transform([predicted_class_id])[0]
            confidence = torch.softmax(logits, dim=1)[0][predicted_class_id].item()

        return (
            f"## DIAGNOSTIC REPORT\n\n{predicted_label}",
            {predicted_label: confidence},
            f"### CLINICAL RECOMMENDATION\n\nBased on the provided symptoms, {predicted_label} is the most likely diagnosis. Please consult a healthcare professional for confirmation."
        )

    except Exception as e:
        return (
            f"## DIAGNOSTIC REPORT\n\n\u274c System Error: {str(e)}",
            {"Error": 0.0},
            "### CLINICAL RECOMMENDATION\n\n*Unable to generate recommendations*"
        )

free_port = find_free_port()

with gr.Blocks(
    title="MediScan AI - Clinical Symptom Analyzer",
    theme=med_theme,
    css="""
    .diagnosis-box {border-left: 4px solid #1a73e8; padding-left: 15px;}
    footer {visibility: hidden}
    .input-box textarea {font-size: 14px !important;}
    .medical-icon {max-width: 80px; margin: 0 auto;}
    """
) as demo:

    gr.Markdown("""
    <div style='text-align: center; margin-bottom: 20px;'>
        <h1 style='font-weight: 600; color: #0d47a1; margin-bottom: 5px;'>
            <i class='fas fa-stethoscope'></i> MediScan AI
        </h1>
        <p style='color: #555;'>Clinical Symptom Analysis Platform | v2.1.4</p>
    </div>
    """)

    with gr.Row(equal_height=True):
        with gr.Column(scale=1):
            gr.HTML("""
            <div style="text-align: center;">
                <svg class="medical-icon" width="80" height="80" viewBox="0 0 24 24" fill="#0d47a1">
                    <path d="M19 3H5c-1.1 0-2 .9-2 2v14c0 1.1.9 2 2 2h14c1.1 0 2-.9 2-2V5c0-1.1-.9-2-2-2zm-2 8h-4v4h-2v-4H7v-2h4V5h2v4h4v2z"/>
                </svg>
            </div>
            """)

            input_box = gr.Textbox(
                label="PATIENT SYMPTOM INPUT",
                placeholder="Enter symptom details (e.g., 'Sharp headache in temples lasting 2 days with nausea')",
                lines=4,
                max_lines=6
            )

            with gr.Row():
                submit_btn = gr.Button("Analyze Symptoms", variant="primary", size="lg")
                clear_btn = gr.Button("Clear", size="lg")

            gr.Examples(
                examples=[
                    ["Persistent headache with photophobia and neck stiffness"],
                    ["Acute chest pain radiating to left arm with sweating"],
                    ["High fever (39.5\u00b0C+) with dry cough and fatigue"]
                ],
                inputs=input_box,
                label="CLINICAL EXAMPLE CASES",
                examples_per_page=3
            )

        with gr.Column(scale=1):
            with gr.Group():
                output_diagnosis = gr.Markdown(
                    value="## DIAGNOSTIC REPORT\n\n*Awaiting analysis...*",
                    elem_classes=["diagnosis-box"]
                )

                confidence_meter = gr.Label(
                    label="CONFIDENCE METRIC",
                    value={"Preliminary": 0.0},
                    num_top_classes=1,
                    show_label=False
                )

                output_recommendation = gr.Markdown(
                    "### CLINICAL RECOMMENDATION\n\n*Analysis pending*"
                )

    gr.HTML("""
    <div style='background: #f8f9fa; padding: 15px; border-radius: 8px; margin-top: 20px;'>
        <h4 style='color: #d32f2f; margin-bottom: 10px;'><i class='fas fa-exclamation-triangle'></i> IMPORTANT DISCLAIMER</h4>
        <p style='color: #555; font-size: 14px;'>
        This AI diagnostic tool provides <b>preliminary analysis only</b> and should not be used as a substitute for professional medical evaluation.
        Always consult with a licensed healthcare provider for clinical decisions. Emergency symptoms (chest pain, difficulty breathing,
        severe trauma) require <b>immediate medical attention</b>.
        </p>
    </div>
    """)

    def update_output(symptom_text):
        result = predict(symptom_text)
        return {
            output_diagnosis: result[0],
            confidence_meter: result[1],
            output_recommendation: result[2]
        }

    submit_btn.click(
        fn=update_output,
        inputs=input_box,
        outputs=[output_diagnosis, confidence_meter, output_recommendation]
    )

    clear_btn.click(
        fn=lambda: ["", {"Preliminary": 0.0}, "### CLINICAL RECOMMENDATION\n\n*Analysis pending*"],
        outputs=[input_box, confidence_meter, output_recommendation]
    )

# Launch the interface
demo.launch(
    share=True,
    server_name="0.0.0.0",
    server_port=free_port,
    show_error=True
)
```
